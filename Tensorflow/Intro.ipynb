```python
normalization_layer = tf.keras.layers.Normalization()

hidden_layer1 = tf.keras.layers.Dense(30, activation= "relu")
hidden_layer2 = tf.keras.layers.Dense(30, activation= "relu")
concat_layer  = tf.keras.layers.Concatenate()
output_layer  = tf.keras.layers.Dense(1)
# ----------------------------------------------------------
input_ = tf.keras.layers.Input(shape= X_train.shape[1:])
normalized = normalization_layer(input_)
hidden1 = hidden_layer1(normalized)
hidden2 = hidden_layer2(hidden1)
concat = concat_layer([normalized, hidden2])
output = output_layer(concat)

model = tf.keras.Model(inputs= [input_], outputs= [output])
```

### Handling multiple inputs

```python
input_wide = tf.keras.layers.Input(shape=[5])   # 0–4-cü xüsusiyyətlər
input_deep = tf.keras.layers.Input(shape=[6])   # 2–7-ci xüsusiyyətlər

norm_wide = tf.keras.layers.Normalization()(input_wide)
norm_deep = tf.keras.layers.Normalization()(input_deep)


hidden1 = tf.keras.layers.Dense(30, activation="relu")(norm_deep)
hidden2 = tf.keras.layers.Dense(30, activation="relu")(hidden1)

concat = tf.keras.layers.concatenate([norm_wide, hidden2])
output = tf.keras.layers.Dense(1)(concat)

model = tf.keras.Model(inputs=[input_wide, input_deep], outputs=[output])

# ----------------------------------------------------------------------

optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
model.compile(loss="mse", optimizer=optimizer, metrics=["RootMeanSquaredError"])

X_train_wide, X_train_deep  = X_train[:, :5],  X_train[:, 2:] # 0:4 , 2:7
X_valid_wide, X_valid_deep  = X_valid[:, :5],  X_valid[:, 2:]   # 0:4 , 2:7
X_test_wide,  X_test_deep   = X_test[:, :5] ,  X_test[:, 2:]   # 0:4 , 2:7
#  use prediction  for  the first  3 samples 
X_new_wide, X_new_deep     = X_test_wide[:3], X_test_deep[:3] # 0:2 , 0:2

tf.keras.layers.Normalization().adapt(X_train_wide)
tf.keras.layers.Normalization().adapt(X_train_deep)

history = model.fit(
                      (X_train_wide, X_train_deep),
                      y_train,
                      epochs=20,
                      validation_data=((X_valid_wide, X_valid_deep), y_valid)
                    )

mse_test = model.evaluate((X_test_wide, X_test_deep),
                           y_test)
y_pred   = model.predict((X_new_wide, X_new_deep))
```

### Handling multiple outputs,
in this example to add an auxiliary output for regularization

```python
# 1. GİRİŞ LAYER-LƏRİ
input_wide = tf.keras.layers.Input(shape=[5], name="input_wide")   # 0–4-cü xüsusiyyətlər
input_deep = tf.keras.layers.Input(shape=[6], name="input_deep")   # 2–7-ci xüsusiyyətlər

# 2. NORMALİZASİYA LAYER-LƏRİNİ ƏVVƏLCƏDƏN YARAT
norm_layer_wide = tf.keras.layers.Normalization(name="norm_wide")
norm_layer_deep = tf.keras.layers.Normalization(name="norm_deep")

# 3. NORMALİZASİYA EDİLMİŞ ÇIXIŞLAR
norm_wide = norm_layer_wide(input_wide)
norm_deep = norm_layer_deep(input_deep)

# 4. DEEP HİSSƏ
hidden1 = tf.keras.layers.Dense(30, activation="relu", name="hidden1")(norm_deep)
hidden2 = tf.keras.layers.Dense(30, activation="relu", name="hidden2")(hidden1)

# 5. BİRLƏŞMƏ
concat = tf.keras.layers.concatenate([norm_wide, hidden2], name="concat")

# 6. ÇIXIŞ LAYER-LƏRİ
output     = tf.keras.layers.Dense(1, name="main_output")(concat)
aux_output = tf.keras.layers.Dense(1, name="aux_output")(hidden2)

# 7. MODEL
model = tf.keras.Model(
    inputs=[input_wide, input_deep],
    outputs=[output, aux_output],
    name="wide_deep_multiple_output"
)

# ---------------------------------------------------------
# 8. OPTIMIZER VƏ COMPILE
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)

model.compile(
                loss={"main_output": "mse", "aux_output": "mse"},
                loss_weights={"main_output": 0.9, "aux_output": 0.1},
                optimizer=optimizer,
                metrics={"main_output": ["RootMeanSquaredError"],
                        "aux_output": ["RootMeanSquaredError"]}
            )

# ---------------------------------------------------------
# 9. NORMALİZASİYANIN ÖYRƏDİLMƏSİ (ADAPT)
# BURASI ƏHƏMİYYƏTLİ: Modeldəki layer-ləri adapt et!
norm_layer_wide.adapt(X_train_wide)
norm_layer_deep.adapt(X_train_deep)

# ---------------------------------------------------------
# 10. MODEL TƏLİMİ
history = model.fit(
                    {"input_wide": X_train_wide, "input_deep": X_train_deep},
                    {"main_output": y_train,     "aux_output": y_train},
                    epochs=20,
                    validation_data=(
                        {"input_wide": X_valid_wide, "input_deep": X_valid_deep},
                        {"main_output": y_valid,     "aux_output": y_valid}
                    ),
                    verbose=0  # Proqress barı göstər
                )

# ---------------------------------------------------------
# 11. MODEL TESTİ
results = model.evaluate(
                        {"input_wide": X_test_wide, "input_deep": X_test_deep},
                        {"main_output": y_test,      "aux_output": y_test},
                        return_dict=True,
                        verbose=0
                    )

print("\nTest nəticələri:")
for key, value in results.items():
    print(f"{key}: {value:.4f}")

# ---------------------------------------------------------
# 12. PROQNOZ
y_pred_main, y_pred_aux = model.predict(
    {"input_wide": X_new_wide, "input_deep": X_new_deep},
    verbose=0
)

print(f"\nİlk 3 proqnoz:")
print(f"Əsas çıxış:  {y_pred_main.flatten()}")
print(f"Köməkçi çıxış: {y_pred_aux.flatten()[:3]}")
```

## Using the Subclassing API to Build Dynamic Models

```python
class WideAndDeepModel(tf.keras.Model):

    def __init__(self, units=30, activation="relu", **kwargs):
        super().__init__(**kwargs)
        self.norm_layer_wide = tf.keras.layers.Normalization()
        self.norm_layer_deep = tf.keras.layers.Normalization()

        self.hidden1 = tf.keras.layers.Dense(units, activation=activation)
        self.hidden2 = tf.keras.layers.Dense(units, activation=activation)
        
        self.main_output = tf.keras.layers.Dense(1)
        self.aux_output  = tf.keras.layers.Dense(1)

    def call(self, inputs):
        input_wide, input_deep = inputs

        norm_wide = self.norm_layer_wide(input_wide)
        norm_deep = self.norm_layer_deep(input_deep)

        hidden1 = self.hidden1(norm_deep)
        hidden2 = self.hidden2(hidden1)

        concat = tf.keras.layers.concatenate([norm_wide, hidden2])

        output = self.main_output(concat)
        aux_output = self.aux_output(hidden2)

        return output, aux_output



# Model instance yarat
model = WideAndDeepModel(units=30, activation="relu", name="my_cool_model")

# Compile et
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss=["mse", "mse"],
    loss_weights=[0.9, 0.1]
)

# Normalizasiya layer-lərini adapt et
model.norm_layer_wide.adapt(X_train_wide)
model.norm_layer_deep.adapt(X_train_deep)

# Təlim et
history = model.fit(
    [X_train_wide, X_train_deep],
    [y_train, y_train],
    epochs=20,
    validation_data=([X_valid_wide, X_valid_deep], [y_valid, y_valid]),
    verbose=0
)

# Proqnoz ver
predictions = model.predict([X_test_wide, X_test_deep])
```

Your Own Callback

```python
class MyCustomCallback(tf.keras.callbacks.Callback):
    def on_train_begin(self, logs=None):
        print("🚀 Təlim başladı!")
    
    def on_train_end(self, logs=None):
        print("✅ Təlim bitdi!")
    
    def on_epoch_begin(self, epoch, logs=None):
        print(f"📊 Epoch {epoch+1} başladı...")
    
    def on_epoch_end(self, epoch, logs=None):
        loss = logs.get('loss', 0)
        val_loss = logs.get('val_loss', 0)
        
        if val_loss > 0:
            ratio = val_loss / loss
            print(f"Epoch {epoch+1}: val/train = {ratio:.2f}")
            
            # Overfitting yoxlama
            if ratio > 2.0:
                print("⚠️  Overfitting ola bilər!")
    
    def on_batch_end(self, batch, logs=None):
        # Hər 100 batch-də bir məlumat ver
        if batch % 100 == 0:
            loss = logs.get('loss', 0)
            print(f"   Batch {batch}: loss = {loss:.4f}")

# model.fit() çağırılanda:
history = model.fit(X_train, y_train, epochs=10, callbacks=[MyCustomCallback()])


# class CompleteCallback(tf.keras.callbacks.Callback):
#     def on_train_begin(self, logs=None):
#         """Təlim BAŞLAYANDA bir dəfə"""
    
#     def on_train_end(self, logs=None):
#         """Təlim BİTƏNDƏ bir dəfə"""
    
#     def on_epoch_begin(self, epoch, logs=None):
#         """Hər epoch BAŞLAYANDA"""
    
#     def on_epoch_end(self, epoch, logs=None):
#         """Hər epoch BİTƏNDƏ"""
    
#     def on_batch_begin(self, batch, logs=None):
#         """Hər batch BAŞLAYANDA"""
    
#     def on_batch_end(self, batch, logs=None):
#         """Hər batch BİTƏNDƏ"""
    
#     # Test/prediction üçün:
#     def on_test_begin(self, logs=None):
#         """model.evaluate() BAŞLAYANDA"""
    
#     def on_test_end(self, logs=None):
#         """model.evaluate() BİTƏNDƏ"""
    
#     def on_predict_begin(self, logs=None):
#         """model.predict() BAŞLAYANDA"""
    
#     def on_predict_end(self, logs=None):
#         """model.predict() BİTƏNDƏ"""
```

```python
# 1. Log qovluğunu yarat
test_logdir = get_run_logdir()
# test_logdir = "my_logs/run_2023_12_25_14_30_00" kimi olacaq

# 2. Yazıçı yarat (bu fayllara yazacaq)
writer = tf.summary.create_file_writer(str(test_logdir))

# 3. Müxtəlif məlumatları yaz
with writer.as_default():  # Bu writer-ı default edir
    for step in range(1, 1000 + 1):  # 1000 addım
        # A) SKALAR - Tək dəyər (qrafik üçün)
        # Sinus dalğası: step artdıqca dəyər dəyişir
        tf.summary.scalar("my_scalar", np.sin(step / 10), step=step)
        # TensorBoard-da: SCALARS → my_scalar
        
        # B) HISTOGRAM - Paylanma
        # Məlumat getdikcə böyüyür (step/100 ilə)
        data = (np.random.randn(100) + 2) * step / 100
        tf.summary.histogram("my_hist", data, buckets=50, step=step)
        # TensorBoard-da: HISTOGRAMS → my_hist
        
        # C) ŞƏKİL - 2 təsadüfi şəkil
        # Şəkillər getdikcə parlaqlaşır (step/1000 ilə)
        images = np.random.rand(2, 32, 32, 3) * step / 1000
        # Format: [sayı, hündürlük, en, kanal]
        tf.summary.image("my_images", images, step=step)
        # TensorBoard-da: IMAGES → my_images
        
        # D) MƏTN - Siyahı formatında
        texts = ["Addım " + str(step), "Onun kvadratı " + str(step ** 2)]
        tf.summary.text("my_text", texts, step=step)
        # TensorBoard-da: TEXT → my_text
        
        # E) AUDIO - Sinus dalğası
        # Tezlik step ilə dəyişir
        sine_wave = tf.math.sin(tf.range(12000) / 48000 * 2 * np.pi * step)
        audio = tf.reshape(tf.cast(sine_wave, tf.float32), [1, -1, 1])
        # Format: [batch_size, audio_length, channels]
        tf.summary.audio("my_audio", audio, sample_rate=48000, step=step)
        # TensorBoard-da: AUDIO → my_audio
```

## Fine-Tuning Neural Network Hyperparameters

```python
import tensorflow as tf
import keras_tuner as kt
import numpy as np

# 1. Məlumatları yüklə
fashion_mnist = tf.keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# 2. Normalizasiya et
X_train = X_train / 255.0
X_test = X_test / 255.0

# 3. Validation set yarat
X_valid, X_train = X_train[:5000], X_train[5000:]
y_valid, y_train = y_train[:5000], y_train[5000:]
```

+ Keras Tuner ( new and fast)

```python
import keras_tuner as kt

def build_model(hp):
    """
    hp = HyperParameters obyekti
    Modeli dinamik qururuq - parametrlər dəyişə bilər
    """
    
    # 1. Hiperparametrləri təyin et
    n_hidden = hp.Int("n_hidden", min_value=0, max_value=8, default=2)
    # 0-8 arası, default 2 gizli qat
    
    n_neurons = hp.Int("n_neurons", min_value=16, max_value=256)
    # Hər qatda 16-256 neyron
    
    learning_rate = hp.Float("learning_rate", 
                            min_value=1e-4,  # 0.0001
                            max_value=1e-2,  # 0.01
                            sampling="log")  # Log miqyasında
    # 0.0001, 0.0002, 0.0005, 0.001, 0.002, 0.005, 0.01
    
    optimizer = hp.Choice("optimizer", values=["sgd", "adam"])
    # Ya 'sgd', ya da 'adam' seçəcək

    #------------------------------------------------------------------------

    # 2. Optimizator seçimi
    if optimizer == "sgd":
        opt = tf.keras.optimizers.SGD(learning_rate=learning_rate)
    else:
        opt = tf.keras.optimizers.Adam(learning_rate=learning_rate)

    #------------------------------------------------------------------------

    # 3. Model quruluşu
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Flatten())  # Giriş qatı
    
    # Gizli qatlar (sayı dəyişir)
    for _ in range(n_hidden):
        model.add(tf.keras.layers.Dense(n_neurons, activation="relu"))
    
    # Çıxış qatı (10 sinif üçün)
    model.add(tf.keras.layers.Dense(10, activation="softmax"))
    
     #-------------------------------------------------------------------------

    # 4. Modeli kompilyasiya et
    model.compile(
        loss="sparse_categorical_crossentropy",
        optimizer=opt,
        metrics=["accuracy"]
    )
    
    return model
```

+ Create Random Search  Tuner  and calling Search ()

```python
# Random Search Tuner yarat
random_search_tuner = kt.RandomSearch(
                build_model,                  # Model qurma funksiyası
                objective="val_accuracy",     # Optimallaşdırılacaq metrik
                max_trials=5,                 # 5 fərqli kombinasiya sınayacaq
                overwrite=True,               # Köhnə nəticələri sil
                directory="my_fashion_mnist", # Nəticələri saxlayacaq qovluq
                project_name="my_rnd_search", # Layihə adı
                seed=42                       # Təkrar oluna bilən nəticələr
            )


# Tənzimləməni başlat
random_search_tuner.search(
            X_train, y_train,
            epochs=10,                    # Hər model 10 epoch təlim olunacaq
            validation_data=(X_valid, y_valid),
            verbose=1                     # Proqressi göstər
        )
```

+ kt.HyperModel

Fine-tune data preprocessing hyperparameters, or model.fit() arguments


```python
class MyClassificationHyperModel(kt.HyperModel):
    # Model qururuq
    def build(self, hp):
        return build_model(hp)  # hp = hyperparameters
    
    # Modeli təlim edirik
    def fit(self, hp, model, X, y, **kwargs):
        # Normalizasiya edib-etməməyi SEÇİR
        if hp.Boolean("normalize"):  # True/False
            norm_layer = tf.keras.layers.Normalization()
            X = norm_layer(X)  # Normalizasiya et
        return model.fit(X, y, **kwargs)
    
    
hyperband_tuner = kt.Hyperband(
                MyClassificationHyperModel(),  # Yuxarıdaki class
                objective="val_accuracy",      # Nəyi maksimum etmək?
                seed = 42,
                max_epochs=10,                 # Max epoch sayı
                hyperband_iterations=2,        #  Bütün proses 2 dəfə təkrarlanır
                factor=3,                      # Hyperband parametri
                overwrite=True,
                directory="my_fashion_mnist",  # Nəticələri harada saxla?
                project_name="hyperband"       # Layihə adı
            )


 
# TensorBoard callback

root_logdir = Path(hyperband_tuner.project_dir) / "tensorboard"
tensorboard_cb = tf.keras.callbacks.TensorBoard(root_logdir)
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=2)

hyperband_tuner.search(
                        X_train,
                          y_train,
                            epochs=10,
                        validation_data=(X_valid, y_valid),
                        callbacks=[early_stopping_cb, tensorboard_cb]
                    )


```
