```python
# PROBLEM  : Threshold kimi parametrlər yadda saxlanılmır

def create_huber(threshold = 1.0):
    
    def huber_fn(y_true,y_pred):
        error = y_true - y_pred
        is_small_error = tf.abs(error) < threshold 
        squared_loss   = tf.abs(error) / 2
        linear_loss    = threshold * tf.abs(error) - threshold **2 / 2 
    
        return tf.where(is_small_error ,squared_loss , linear_loss)
    
    return  huber_fn

model.compile(loss = create_huber(2.0), optimizer = 'nadam')
```

```python
# HƏLL YOLU: Loss SUBCLASS-ı YARATMAQ

class HuberLoss(tf.keras.losses.Loss):
 
 def __init__(self, threshold=1.0, **kwargs):
    self.threshold = threshold
    super().__init__(**kwargs)
 
 def call(self, y_true, y_pred):
    error = y_true - y_pred
    is_small_error = tf.abs(error) < self.threshold
    squared_loss = tf.square(error) / 2
    linear_loss = self.threshold * tf.abs(error) - self.threshold**2 / 2

    return tf.where(is_small_error, squared_loss, linear_loss)
 
 def get_config(self):
    base_config = super().get_config()
    print("base_config:", base_config)  # GÖRÜN NƏ QAYIDIR?
    return {**base_config, "threshold": self.threshold}
 
#  base config-in yoxlamasi 

loss = HuberLoss(  threshold=2.0, 
                 # reduction='sum',
                   name='my_huber'
                   )
config = loss.get_config()
print("Yekun config:", config)
```

```python
# Xüsusi aktivasiya funksiyası (softplus-a bənzər)
def my_softplus(z):
    return tf.math.log(1.0 + tf.exp(z))

# Xüsusi Glorot inisializatoru
def my_glorot_initializer(shape, dtype=tf.float32):
    stddev = tf.sqrt(2. / (shape[0] + shape[1]))
    return tf.random.normal(shape, stddev=stddev, dtype=dtype)

# Xüsusi L1 requlyarizatoru
def my_l1_regularizer(weights):
    return tf.reduce_sum(tf.abs(0.01 * weights))

# Xüsusi konstraint (bütün çəkilər müsbət olmalı)
def my_positive_weights(weights):
    return tf.where(weights < 0., tf.zeros_like(weights), weights)

layer = tf.keras.layers.Dense(
                        1,
                        activation        = my_softplus,           # Aktivasiya
                        kernel_initializer= my_glorot_initializer, # İnisializator
                        kernel_regularizer= my_l1_regularizer,     # Requlyarizator
                        kernel_constraint = my_positive_weights    # Konstraint
                    )
```

Custom Streaming Metric yaratmaq

```python
class HuberMetric(tf.keras.metrics.Metric):
    def __init__(self, threshold=1.0, **kwargs):
        super().__init__(**kwargs)  # Ana klasın parametrlərini işlə
        self.threshold = threshold   # Threshold dəyərini saxla
        self.huber_fn = create_huber(threshold)  # Huber funksiyasını yarat
        self.total = self.add_weight("total", initializer="zeros") # BÜTÜN itkilərin cəmi (əvvəl 0)
        self.count = self.add_weight("count", initializer="zeros") # NÜMUNƏLƏRİN sayı (əvvəl 0)

    def update_state(self, y_true, y_pred, sample_weight=None):
        sample_metrics = self.huber_fn(y_true, y_pred)  # Hər nümunə üçün itki
        
        # Batch-dəki itkiləri cəmlə və ümumi cəmə əlavə et
        self.total.assign_add(tf.reduce_sum(sample_metrics))
        # Batch-dəki nümunə sayını götür və ümumi sayı artır
        self.count.assign_add(tf.cast(tf.size(y_true), tf.float32))
    
    def result(self):
        return self.total / self.count  # ORTALAMA itki = cəm / say
    
    def get_config(self):
        base_config = super().get_config()  # Ana klasın konfiqini al
        return {**base_config, "threshold": self.threshold}  # Öz parameteri də əlavə et
```

#### Stateful (weight-li) Custom Layer yaratmaq

```python
class MyDense(tf.keras.layers.Layer):
    # 1️⃣ CONSTRUCTOR (__init__) - Layer-i yaradırıq
    def __init__(self, units, activation=None, **kwargs):
        super().__init__(**kwargs)        # Ana klasın parametrlərini ötür
        self.units = units                 # Neçə neyron? (10, 20, 100...)
        # Activation-ı qəbul et (string, funksiya və ya None)
        self.activation = tf.keras.activations.get(activation)

    # 2️⃣ build() - Çəkiləri yaradır (input ölçüsü bəlli olandan sonra)
    def build(self, batch_input_shape):
        # KERNEL - girişlərlə neyronlar arasındakı çəkilər
        # Şəkil: [giriş_sayı, neyron_sayı]
        self.kernel = self.add_weight(
                name="kernel",
                shape=[batch_input_shape[-1], self.units],
                initializer="glorot_normal")   # İlkin dəyər metodu

        # BIAS - hər neyronun qərəzi
        # Şəkil: [neyron_sayı]
        self.bias = self.add_weight(
                        name="bias",
                        shape=[self.units],
                        initializer="zeros")  # Qərəzlər 0-dan başlayır

    # 3️⃣ call() - İrəli hesablama (forward pass)
    def call(self, X):
        # X @ kernel + bias  =  giriş * çəkilər + qərəz
        # Sonra activation-dan keçir (ReLU, sigmoid və s.)
        return self.activation(X @ self.kernel + self.bias)
    
    # Lakin dynamic layer varsa (məsələn, ölçü dəyişir)
    def compute_output_shape(self, input_shape):
        return (input_shape[0], self.units)  # batch_size, units

    # 4️⃣ get_config() - Yaddaş üçün parametrləri saxla
    def get_config(self):
        base_config = super().get_config()      # Ana klasın konfiqi
        return {
            **base_config,                       # Hamısını birləşdir
            "units": self.units,                  # Öz parametrimiz
            "activation": tf.keras.activations.serialize(self.activation)  # Activation adı
        }

```

```python
# Çox Girişli / Çox Çıxışlı Layer

class MyMultiLayer(tf.keras.layers.Layer):
    def call(self, X):
        X1, X2 = X                    # 2 giriş gəlir
        çıxış1 = X1 + X2               # Toplama
        çıxış2 = X1 * X2               # Vurma  
        çıxış3 = X1 / X2               # Bölmə
        return çıxış1, çıxış2, çıxış3  # 3 çıxış qaytarır
```

```python
# Training/Testing Fərqli Davranış

class MyGaussianNoise(tf.keras.layers.Layer):
    def __init__(self, stddev, **kwargs):
        super().__init__(**kwargs)
        self.stddev = stddev
    
    def call(self, X, training=False):
        if training:
            # TRAINING: səs əlavə et
            noise = tf.random.normal(tf.shape(X), stddev=self.stddev)
            return X + noise
        else:
            # TESTING: təmiz X
            return X
```

### Custom Models 

```python
# Subclassing API  yanasmasi

class MyModel(tf.keras.Model):
    def __init__(self):
        super().__init__()
        # Layer-ləri burada yarat
        self.dense1 = tf.keras.layers.Dense(10)
    
    def call(self, inputs):
        # Forward pass məntiqi
        return self.dense1(inputs)
```

```python
#  Residual  bloku
class ResidualBlock(tf.keras.layers.Layer):
    def __init__(self, n_layers, n_neurons):
        super().__init__()
        # İçində bir neçə Dense layer saxlayır
        self.hidden = [tf.keras.layers.Dense(n_neurons, activation="relu") 
                       for _ in range(n_layers)]
    
    def call(self, inputs):
        Z = inputs
        for layer in self.hidden:  # Bütün layer-lərdən keç
            Z = layer(Z)
        return inputs + Z  # Skip connection (girişi əlavə et)
```

```python
# Tam sekilde 

class ResidualRegressor(tf.keras.Model):
    def __init__(self, output_dim):
        super().__init__()
        self.hidden1 = tf.keras.layers.Dense(30, activation="relu")
        self.block1 = ResidualBlock(2, 30)  # 2 layer-li blok
        self.block2 = ResidualBlock(2, 30)  # 2 layer-li blok
        self.out = tf.keras.layers.Dense(output_dim)
    
    def call(self, inputs):
        Z = self.hidden1(inputs)
        # Eyni bloku 4 dəfə təkrarla
        for _ in range(4):
            Z = self.block1(Z)
        Z = self.block2(Z)  # Başqa blok
        return self.out(Z)
```

### Model Daxili Loss və Metric

```python
class ReconstructingRegressor(tf.keras.Model):
    # ========== 1. CONSTRUCTOR ==========
    def __init__(self, output_dim):
        super().__init__()
        # 5 gizli təbəqə - hər biri 30 neyronlu
        self.hidden = [tf.keras.layers.Dense(30, activation="relu") 
                       for _ in range(5)]
        # Əsas çıxış təbəqəsi (regresiya üçün)
        self.out = tf.keras.layers.Dense(output_dim)
        # Reconstruction error-u izləmək üçün METRIC
        # Mean() = bütün batch-lərin ortalamasını saxlayır
        self.reconstruction_mean = tf.keras.metrics.Mean(name="reconstruction_error")
    
    # ========== 2. BUILD ==========
    def build(self, batch_input_shape):
        # batch_input_shape = (batch_size, input_feature_sayı)
        # Məsələn: (None, 64) -> 64 feature
        n_inputs = batch_input_shape[-1]  # input_feature_sayı = 64
        # Reconstruction layer - input-u bərpa etmək üçün
        # Çıxış ölçüsü input-la EYNİ olmalıdır!
        self.reconstruct = tf.keras.layers.Dense(n_inputs)
    
    # ========== 3. CALL (FORWARD PASS) ==========
    def call(self, inputs, training=False):
        # inputs = giriş məlumatları
        # ----- Gizli təbəqələrdən keçid -----
        Z = inputs
        for layer in self.hidden:
            Z = layer(Z)  # Z = gizli reprezentasiya
        # ----- INPUT-U BƏRPA ET -----
        reconstruction = self.reconstruct(Z)  # Gizli reprezentasiyadan input-u bərpa et
        # ----- RECONSTRUCTION LOSS -----
        # MSE: (reconstruction - inputs)^2 ortalaması
        recon_loss = tf.reduce_mean(tf.square(reconstruction - inputs))
        # ----- LOSS-U MODELƏ ƏLAVƏ ET -----
        # 0.05 = çəki faktoru (əsas loss-u basmasın deyə)
        self.add_loss(0.05 * recon_loss)  # Bu loss avtomatik ümumi loss-a əlavə olunur
        # ----- TRAINING ZAMANI METRIC-I YENİLƏ -----
        if training:
            # reconstruction_error metric-inə dəyər əlavə et
            self.add_metric(recon_loss, name="reconstruction_error")
        # ----- ƏSAS ÇIXIŞ (REGRESİYA) -----
        return self.out(Z)
```

```python
#  Custom gradient yarat

@tf.custom_gradient
def my_softplus(z):
    # Əsas funksiya (forward pass)
    result = tf.math.log(1 + tf.exp(-tf.abs(z))) + tf.maximum(0., z)
    
    # Xüsusi gradient funksiyası (backward pass)
    def my_softplus_gradients(grads):
        # grads = yuxarıdan gələn gradient
        return grads * (1 - 1/(1 + tf.exp(z)))  # Stabil gradient
        
    return result, my_softplus_gradients
```

### Custom Training Loops

```python
# ========== 1. MODEL (L2 regularizer ilə) ==========
l2_reg = tf.keras.regularizers.l2(0.05)  # 0.05 = regularization gücü

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(30, activation="relu", 
                          kernel_initializer="he_normal",
                          kernel_regularizer=l2_reg),  # L2 regularization
    tf.keras.layers.Dense(1, kernel_regularizer=l2_reg)
])
```

```python
# ========== 2. KÖMƏKÇİ FUNKSİYALAR ==========
def random_batch(X, y, batch_size=32):
    """Dataset-dən təsadüfi batch seçir"""
    idx = np.random.randint(len(X), size=batch_size)
    return X[idx], y[idx]

def print_status_bar(step, total, loss, metrics=None):
    """Təlim statusunu göstərir (eyni sətirdə yenilənir)"""
    metrics_str = " - ".join([f"{m.name}: {m.result():.4f}" 
                              for m in [loss] + (metrics or [])])
    end = "" if step < total else "\n"
    print(f"\r{step}/{total} - {metrics_str}", end=end)
```

```python
# ========== 3. HYPERPARAMETRLƏR ==========
n_epochs = 5
batch_size = 32
n_steps = len(X_train) // batch_size  # Hər epoch-da neçə batch

optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
loss_fn = tf.keras.losses.mean_squared_error

# Metric-lər (streaming - yığılır)
mean_loss = tf.keras.metrics.Mean(name="mean_loss")
metrics = [tf.keras.metrics.MeanAbsoluteError()]
```

```python
# ========== 4. ƏSAS TRAIN LOOP ==========
for epoch in range(1, n_epochs + 1):
    print(f"Epoch {epoch}/{n_epochs}")
    
    for step in range(1, n_steps + 1):
        # Batch seç
        X_batch, y_batch = random_batch(X_train_scaled, y_train)
        
        # Gradient hesabla
        with tf.GradientTape() as tape:
            y_pred = model(X_batch, training=True)  # Forward pass
            main_loss = tf.reduce_mean(loss_fn(y_batch, y_pred))  # MSE
            # ƏSAS: main_loss + bütün regularization loss-ları
            loss = tf.add_n([main_loss] + model.losses) # model.losses = L2 loss
        
        # Gradient-ləri hesabla və tətbiq et
        gradients = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        
        # Metric-ləri yenilə
        mean_loss(loss)  # Loss-u yığ
        for metric in metrics:
            metric(y_batch, y_pred)  # MAE-ni yığ
        
        # Status göstər
        print_status_bar(step, n_steps, mean_loss, metrics)
    
    # Hər epoch sonu metric-ləri SIFIRLA (yeni epoch üçün)
    for metric in [mean_loss] + metrics:
        metric.reset_states()
```

### AutoGraph və Tracing 

```python
#  Sizin yazdığınız:

import tensorflow as tf

# Adi Python funksiyası
@tf.function
def sum_squares(x):
    total = 0
    for i in range(x):  # Python for loop
        total += i ** 2
    return total

# Tracing merheelsi 
result = sum_squares(tf.constant(10))
print(result)  # 30 (0²+1²+2²+3²+4²)

# AutoGraph-un yaratdığı koda baxaq
print(tf.autograph.to_code(sum_squares.python_function))
```

```python
# AutoGraph çevirir:

def tf_sum_squares(x):
    total = 0
    
    # loop_body = hər addımda nə olacaq
    def loop_body(i):
        nonlocal total
        total += i ** 2
    
    # for loop-u TF while_loop-a çevir
    tf.for_stmt(
        loop_vars=[total],
        loop_body=loop_body,
        loop_range=range(x)
    )
    
    return total
```

```python
# Yaradılan Qrafikə Necə Baxmaq Olar?

import tensorflow as tf
import datetime

# Funksiya
@tf.function
def sum_squares(x):
    total = 0
    for i in range(x):
        total += i ** 2
    return total

# TensorBoard üçün log qovluğu
log_dir = "logs/func/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

# Trace aktiv et
tf.summary.trace_on(graph=True, profiler=True)

# Funksiyanı çağır
result = sum_squares(tf.constant(5))
print(f"Nəticə: {result}")

# Trace-i yaz
with tf.summary.create_file_writer(log_dir).as_default():
    tf.summary.trace_export(
        name="sum_squares",
        step=0,
        profiler_outdir=log_dir
    )

print(f"\nTensorBoard-u işə sal:")
print(f"tensorboard --logdir {log_dir}")
print(f"\nSonra brauzerdə http://localhost:6006 açın")
```
